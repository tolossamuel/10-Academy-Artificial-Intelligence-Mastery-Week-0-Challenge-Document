```python
mean_ghi = benin["GHI"].mean()
midean_ghi = benin["GHI"].median()
print(mean_ghi, midean_ghi)
```

This GHI is positively skewed
that means the mean is greater than midean
mean = 240.55
midean = 1.8

```python

gHI_range = pd.cut(benin['GHI'], bins=range(int(benin['GHI'].min()), int(benin['GHI'].max()) + 101, 100))
ghi_counts = gHI_range.value_counts().sort_index()
ghi_counts.plot(kind='bar', xlabel='GHI Range', ylabel='Count', title='Number of Data Points in GHI Ranges')
```

```python
mean_ghi = benin["DNI"].mean()
midean_ghi = benin["DNI"].median()
print(mean_ghi, midean_ghi)
```
